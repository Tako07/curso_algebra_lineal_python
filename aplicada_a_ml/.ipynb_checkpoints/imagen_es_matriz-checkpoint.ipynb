```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
plt.style.use('classic')

imagen = Image.open('./imagenes/cebra.jpg')

plt.imshow(imagen)
```

```python
#a escala de grises
imagen_gr = imagen.convert('LA')
print(imagen_gr)
```

```python
image_mat = np.array(list(imagen_gr.getdata(band=0)), float)
print(image_mat)
```

```python
image_mat.shape = (imagen_gr.size[1], imagen_gr.size[0])
print(image_mat)
```

```python
print(image_mat.shape)
```

```python
plt.imshow(image_mat, cmap='gray')
plt.show()
```

```python
imagen_mat_2 = image_mat/10
print(imagen_mat_2)
```

```python
plt.imshow(imagen_mat_2)
plt.show()
```

```python
plt.imshow(imagen_mat_2, cmap='gray')
plt.show()
```

```python
print(np.max(imagen_mat_2))
print(np.max(image_mat))
```

```python
print(np.min(imagen_mat_2))
print(np.min(image_mat))
```
