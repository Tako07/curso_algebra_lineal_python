```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace( -5, 5, 1000)
#sistema de ecuaciones
y_1 = -4*x + 3
y_2 = 2*x + 5
y_3 = -3*x + 1

plt.plot(x, y_1)
plt.plot(x, y_2)
plt.plot(x, y_3)

plt.xlim(-2,2.5)
plt.ylim(-6,6)
plt.show()
```

```python
matriz = np.array([[4,1], [-2,3], [3,1]])
matriz_pse = np.linalg.pinv(matriz)
print(matriz_pse)
```

```python
b = np.array([[3],[5],[1]])
print(b)
```

```python
resultado = matriz_pse.dot(b)
print(resultado)
```

```python
plt.plot(x, y_1)
plt.plot(x, y_2)
plt.plot(x, y_3)

plt.xlim(-2,2.5)
plt.ylim(-6,6)

plt.scatter(resultado[0], resultado[1])
plt.show()
```
